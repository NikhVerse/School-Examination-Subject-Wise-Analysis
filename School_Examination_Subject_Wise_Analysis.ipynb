```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# 1. Setup & Data Generation
np.random.seed(42)
num_students = 1000
```

```python
# Generating scores for the requested subjects
data = {
    'Maths': np.random.normal(loc=70, scale=15, size=num_students).clip(0, 100),
    'Science': np.random.normal(loc=75, scale=12, size=num_students).clip(0, 100),
    'Physics': np.random.normal(loc=65, scale=18, size=num_students).clip(0, 100),
    'Chemistry': np.random.normal(loc=68, scale=14, size=num_students).clip(0, 100),
    'Biology': np.random.normal(loc=78, scale=10, size=num_students).clip(0, 100)
}
```

```python
# Create the DataFrame
df = pd.DataFrame(data).round(0).astype(int)
```

```python
# 2. Analysis
subject_stats = df.agg(['mean', 'std', 'min', 'max']).transpose()
subject_stats_sorted = subject_stats.sort_values(by='mean')

print("Subject Statistics (Sorted by Difficulty):")
print(subject_stats_sorted)
```

```python
# 3. Visualization (Default Style)
plt.figure(figsize=(10, 6))

# Melt the data for plotting
df_melted = df.melt(var_name='Subject', value_name='Marks')
```

```python
# Create the Box Plot with default styling
sns.boxplot(data=df_melted, x='Subject', y='Marks')

plt.title('Subject-Wise Performance Analysis')
plt.ylabel('Score Distribution')
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.show()
```

```python
# Automated Conclusion
hardest = subject_stats_sorted.index[0]
easiest = subject_stats_sorted.index[-1]
print(f"\nConclusion: The hardest subject is {hardest}, and the easiest is {easiest}.")
```
